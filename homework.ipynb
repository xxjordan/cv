```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# 读取图片

```python
img_bgr = cv2.imread('./week1_homework.png')
img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

# 进行两次中值滤波
## 中值滤波能有效的去除异常值，不像均值滤波容易受异常值干扰

```python
filter_img = cv2.medianBlur(img_rgb,13)
filter_img = cv2.medianBlur(filter_img,9)
plt.imshow(filter_img)
```

# 打上水印

```python
waterman = cv2.imread('./water.png')
waterman = cv2.resize(waterman,filter_img.shape[:-1][::-1])
waterman = 255 - waterman
water_img = cv2.addWeighted(filter_img,1,waterman,0.5,1)
```

```python
plt.imshow(water_img)
```

# 保存图片

```python
water_img_bgr = cv2.cvtColor(water_img,cv2.COLOR_RGB2BGR)
cv2.imwrite('homework_out.png',water_img_bgr)
```
